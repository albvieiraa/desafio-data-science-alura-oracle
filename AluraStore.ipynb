## 1. Análise do faturamento

Calcular o faturamento total de cada loja. Somando os valores da coluna Preço de cada loja para estimar o faturamento.



```python
faturamento_loja1 = loja1['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

faturamentos = [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

for i in range(len(faturamentos)):
  print(f"Faturamento da {lojas[i]}: R$ {faturamentos[i]:.2f}")
```

```python
# Plotando no gráfico
plt.bar(lojas, faturamentos)
plt.title("Faturamento Total por Loja")
plt.xlabel('Loja')
plt.ylabel('Faturamento (R$)')
plt.show()
```

```python
# Vendas ao longo do tempo Loja 1

# Agrupamento os dados por data e loja, somando o faturamento
faturamento_por_data_loja1 = loja1.groupby('Data da Compra')['Preço'].sum()

# Plotando gráfico
plt.plot(faturamento_por_data_loja1.index, faturamento_por_data_loja1.values, label='Loja 1')
plt.title('Faturamento ao Longo do Tempo - Loja 1')
plt.xlabel('Data')
plt.ylabel('Faturamento (R$)')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
# Vendas ao longo do tempo Loja 2

faturamento_por_data_loja2 = loja2.groupby('Data da Compra')['Preço'].sum()

# Plotando gráfico
plt.plot(faturamento_por_data_loja2.index, faturamento_por_data_loja2.values, label='Loja 2')
plt.title('Faturamento ao Longo do Tempo - Loja 2')
plt.xlabel('Data')
plt.ylabel('Faturamento (R$)')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
# Vendas ao longo do tempo Loja 3

faturamento_por_data_loja3 = loja3.groupby('Data da Compra')['Preço'].sum()

# Plotando gráfico
plt.plot(faturamento_por_data_loja3.index, faturamento_por_data_loja3.values, label='Loja 3')
plt.title('Faturamento ao Longo do Tempo - Loja 3')
plt.xlabel('Data')
plt.ylabel('Faturamento (R$)')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
# Vendas ao longo do tempo Loja 4

faturamento_por_data_loja4 = loja4.groupby('Data da Compra')['Preço'].sum()

# Plotando gráfico
plt.plot(faturamento_por_data_loja4.index, faturamento_por_data_loja4.values, label='Loja 4')
plt.title('Faturamento ao Longo do Tempo - Loja 4')
plt.xlabel('Data')
plt.ylabel('Faturamento (R$)')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

## 2. Vendas por Categoria

Calcular a quantidade de produtos vendidos por categoria em cada loja. A ideia é agrupar os dados por categoria e contar o número de vendas de cada tipo, mostrando as categoria mais populares de cada loja

```python
loja1.columns
```

```python
# Verificando
categorias_loja1 = loja1.groupby('Categoria do Produto')['Produto'].count().sort_values(ascending=False)
categorias_loja2 = loja2.groupby('Categoria do Produto')['Produto'].count().sort_values(ascending=False)
categorias_loja3 = loja3.groupby('Categoria do Produto')['Produto'].count().sort_values(ascending=False)
categorias_loja4 = loja4.groupby('Categoria do Produto')['Produto'].count().sort_values(ascending=False)

categorias = [categorias_loja1, categorias_loja2, categorias_loja3, categorias_loja4]

# Printando
for i, categorias in enumerate(categorias, start=1):
    print(f"\nCategorias mais vendidas - Loja {i}")
    print(categorias)
```

```python
# plotando gráfico

# Função
def plot_categorias(categorias, titulo_loja):
  plt.figure(figsize=(10,6))
  categorias.plot(kind='bar', color='skyblue')
  plt.title(titulo_loja)
  plt.xlabel('Categoria do Produto')
  plt.ylabel('Quantidade de Produtos Vendidos')
  plt.xticks(rotation=45, ha='right')
  plt.tight_layout()
  plt.show()
```

```python
# Plotando os gráficos para cada loja
plot_categorias(categorias_loja1, 'Categorias mais populares da Loja 1')
plot_categorias(categorias_loja2, 'Categorias mais populares da Loja 2')
plot_categorias(categorias_loja3, 'Categorias mais populares da Loja 3')
plot_categorias(categorias_loja4, 'Categorias mais populares da Loja 4')
```

## 3. Média de Avaliação das Lojas

Calcular a média das avaliações dos clientes para cada loja. O objetivo é entender a satisfação dos clientes com os produtos vendidos.

```python
loja1.columns
```

```python
# Calcular a média
media_aval_loja1 = loja1['Avaliação da compra'].mean()
media_aval_loja2 = loja2['Avaliação da compra'].mean()
media_aval_loja3 = loja3['Avaliação da compra'].mean()
media_aval_loja4 = loja4['Avaliação da compra'].mean()

medias_lojas = [media_aval_loja1,media_aval_loja2,media_aval_loja3,media_aval_loja4]

for i in range(len(medias_lojas)):
  print(f"Média de avaliação da {lojas[i]}: {medias_lojas[i]:.2f}")
```

```python
plt.bar(lojas, medias_lojas, color=['skyblue', 'lightcoral', 'gold', 'lightgreen'])
plt.title('Média de Avaliação por Loja')
plt.xlabel('Loja')
plt.ylabel('Média de Avaliação')
plt.ylim(0, 5)
plt.show()
```

```python
# Satisfação x Produto - agrupamento
satisfacao_produto_loja1 = loja1.groupby('Produto')['Avaliação da compra'].mean()
satisfacao_produto_loja2 = loja2.groupby('Produto')['Avaliação da compra'].mean()
satisfacao_produto_loja3 = loja3.groupby('Produto')['Avaliação da compra'].mean()
satisfacao_produto_loja4 = loja4.groupby('Produto')['Avaliação da compra'].mean()

#Ordenação
satisfacao_produto_order1 = satisfacao_produto_loja1.sort_values(ascending=False)
satisfacao_produto_order2 = satisfacao_produto_loja2.sort_values(ascending=False)
satisfacao_produto_order3 = satisfacao_produto_loja3.sort_values(ascending=False)
satisfacao_produto_order4 = satisfacao_produto_loja4.sort_values(ascending=False)

satisfacao_cliente = [satisfacao_produto_order1, satisfacao_produto_order2, satisfacao_produto_order3, satisfacao_produto_order4]

# Printar resultados
for i in range(len(satisfacao_cliente)):
  print(f"\nMédia de Avaliação do Produto da {lojas[i]}:")
  for produto, avaliacao in satisfacao_cliente[i].items():  # itera por cada item
    print(f"{produto}: {avaliacao:.2f}")  # formata a avaliação
```

```python
# Satisfação por categoria

# Agrupamento
satisfacao_categoria_loja1 = loja1.groupby('Categoria do Produto')['Avaliação da compra'].mean()
satisfacao_categoria_loja2 = loja2.groupby('Categoria do Produto')['Avaliação da compra'].mean()
satisfacao_categoria_loja3 = loja3.groupby('Categoria do Produto')['Avaliação da compra'].mean()
satisfacao_categoria_loja4 = loja4.groupby('Categoria do Produto')['Avaliação da compra'].mean()

#Ordenação
satisfacao_categoria_order1 = satisfacao_categoria_loja1.sort_values(ascending=False)
satisfacao_categoria_order2 = satisfacao_categoria_loja2.sort_values(ascending=False)
satisfacao_categoria_order3 = satisfacao_categoria_loja3.sort_values(ascending=False)
satisfacao_categoria_order4 = satisfacao_categoria_loja4.sort_values(ascending=False)

satisfacao_categoria = [satisfacao_categoria_order1, satisfacao_categoria_order2, satisfacao_categoria_order3, satisfacao_categoria_order4]

# Printar resultados
for i in range(len(satisfacao_categoria)):
  print(f"\nMédia de Avaliação do Produto da {lojas[i]}:")
  for categoria, avaliacao in satisfacao_categoria[i].items():  # itera por cada item
    print(f"{categoria}: {avaliacao:.2f}")  # formata a avaliação
```

```python
# Plotando os gráficos para cada loja
for i, satisfacao_categoria_order in enumerate(satisfacao_categoria):
    plt.figure(figsize=(10, 6))
    plt.bar(satisfacao_categoria_order.index, satisfacao_categoria_order.values)
    plt.title(f'Média de Avaliação por Categoria - {lojas[i]}')
    plt.xlabel('Categoria do Produto')
    plt.ylabel('Média de Avaliação')
    plt.xticks(rotation=45, ha='right')
    plt.ylim(0, 5)
    plt.tight_layout()
    plt.show()
```

```python
# Dados para o gráfico
categorias = satisfacao_categoria_order1.index  # Usa as categorias da Loja 1 como referência
num_lojas = len(lojas)
avaliacoes_lojas = [loja.reindex(categorias, fill_value=0) for loja in satisfacao_categoria]  # Reindexa para ter as mesmas categorias em todas as lojas

# Largura das barras
bar_width = 0.2

# Posições das barras no eixo x
x_pos = np.arange(len(categorias))

# Criando o gráfico
fig, ax = plt.subplots(figsize=(12, 6))

for i in range(num_lojas):
  ax.bar(x_pos + i * bar_width, avaliacoes_lojas[i].values, bar_width, label=lojas[i])

# Ajustes no gráfico
ax.set_ylabel('Média de Avaliação')
ax.set_title('Média de Avaliação por Categoria e Loja')
ax.set_xticks(x_pos + bar_width * (num_lojas -1)/ 2)  # Centraliza os rótulos das categorias
ax.set_xticklabels(categorias, rotation=45, ha='right')  # Rotaciona os rótulos das categorias
ax.legend()
plt.tight_layout()
plt.show()
```

A loja 1 teve a menor nota de avaliação dos clientes, em relação aos produtos. Contudo, foi a loja que mais houveram compras então a variação das notas é maior, podendo interferir na média.

A loja 4 obteve pontuação média de 4.0, porém o faturamento foi o mais baixo das 4 lojas.

Além disso, as lojas 2 e 3 obtiveram avaliações semelhantes, assim como o faturamento.

## 4. Produtos Mais e Menos Vendidos

Identificar quais produtos foram os mais vendidos e os menos vendidos em cada loja.  Ao final, visualize os resultados de forma que fique claro quais produtos se destacaram nas vendas de cada loja.

```python
loja1.columns
```

```python
# Agrupando por produto e contando sua quantidade por loja
produtos_vendidos_loja1 = loja1.groupby('Produto')['Produto'].count().sort_values(ascending=False)
produtos_vendidos_loja2 = loja2.groupby('Produto')['Produto'].count().sort_values(ascending=False)
produtos_vendidos_loja3 = loja3.groupby('Produto')['Produto'].count().sort_values(ascending=False)
produtos_vendidos_loja4 = loja4.groupby('Produto')['Produto'].count().sort_values(ascending=False)

# Guardando numa variável
produtos_vendidos_lojas = [produtos_vendidos_loja1, produtos_vendidos_loja2, produtos_vendidos_loja3, produtos_vendidos_loja4]
```

```python
# 5 produtos mais vendidos
mais_vendidos_loja1 = produtos_vendidos_loja1.head(5)
mais_vendidos_loja2 = produtos_vendidos_loja2.head(5)
mais_vendidos_loja3 = produtos_vendidos_loja3.head(5)
mais_vendidos_loja4 = produtos_vendidos_loja4.head(5)

# 5 produtos menos vendidos
menos_vendidos_loja1 = produtos_vendidos_loja1.tail(5)
menos_vendidos_loja2 = produtos_vendidos_loja2.tail(5)
menos_vendidos_loja3 = produtos_vendidos_loja3.tail(5)
menos_vendidos_loja4 = produtos_vendidos_loja4.tail(5)


for i, produtos_loja in enumerate(produtos_vendidos_lojas):
    print(f"{lojas[i]} - Mais Vendidos:")
    print(produtos_loja.head(5))

    print(f"\n{lojas[i]} - Menos Vendidos:")
    print(produtos_loja.tail(5))

    print("\n" + "-"*50 + "\n")
```
